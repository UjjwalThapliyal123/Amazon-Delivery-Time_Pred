```python
#find the distance between store and drop locations
from math import radians, cos, sin, asin, sqrt
```

```python
def calculate_distance(lon1,lat1,lon2,lat2):
  lon1,lat1,lon2,lat2=map(radians,[lon1,lat1,lon2,lat2])

  dist_lon=lon2-lon1
  dist_lat=lat2-lat1
  a=sin(dist_lat/2)**2+cos(lat1)*cos(lat2)*sin(dist_lon/2)**2
  c=2* asin(sqrt(a))
  r=6371 # earh radius in km
  return c*r
```

```python
amazon['distance']=amazon.apply(lambda row: calculate_distance(row['Store_Latitude'],row['Store_Longitude'],row['Drop_Latitude'],row['Drop_Longitude']),axis=1)
```

```python
amazon['Delivery_Time'].head()
```

```python
# Distribution of delivery times.
plt.figure(figsize=(8,7))
sns.boxplot(amazon['Delivery_Time'])
plt.show()
```

```python
# Impact of weather and traffic on delivery time
plt.figure(figsize=(11,7))
sns.lineplot(x=amazon['Delivery_Time'],y=amazon['Weather'],data=amazon,color="green")
sns.lineplot(x=amazon['Delivery_Time'],y=amazon['Traffic'],data=amazon,color="yellow")
plt.xlabel('Delivery_Time')
plt.ylabel('Traffic and Weather')
plt.grid(True)
plt.show()
```

```python
# Relationship between distance and delivery time.
plt.figure(figsize=(11,7))
sns.lineplot(x=amazon['Delivery_Time'],y=amazon['distance'],data=amazon)
plt.xlabel('Delivery_Time')
plt.ylabel('distance')
plt.grid(True)
plt.show()
```

```python
agent_performance=amazon.groupby(['Agent_Rating','Weather','Traffic'])['Delivery_Time'].mean().reset_index()
```

```python
agent_performance.head()
```

```python

# Bar chart for average delivery time by agent and weather
plt.figure(figsize=(12,6))
sns.barplot(x='Agent_Rating',y='Delivery_Time',hue='Weather',data=agent_performance)
plt.title('Average Delivery Time due to  Weather')
plt.xlabel('Agent Rating')
plt.ylabel('Average Delivery Time')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Agent_Rating',y='Delivery_Time',hue='Traffic',data=agent_performance)
plt.title('Average Delivery Time due to Traffic')
plt.xlabel('Agent Rating')
plt.ylabel('Average Delivery Time')
plt.show()
```
